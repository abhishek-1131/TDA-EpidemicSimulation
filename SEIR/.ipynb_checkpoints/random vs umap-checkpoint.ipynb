```python
%%time
import time
import matplotlib.pyplot as plt
import umap.umap_ as umap
from sklearn.datasets import make_blobs
import numpy as np
%matplotlib inline
import scipy
import time
import networkx as nx
import collections
import EoN
from collections import defaultdict
import random
```

```python
s = 500
arr = [int(0.2*s),int(0.05*s),int(0.1*s),int(0.16*s),int(0.12*s),int(0.1*s),int(0.04*s),int(0.08*s),int(0.15*s)]
data,y = make_blobs(n_samples=arr,n_features=2,random_state=16, cluster_std=1)
```

```python
plt.scatter(data[:,0],data[:,1], c = y)
```

```python
Z1 = nx.Graph()

for i in range(len(data)):
    Z1.add_node(i,pos=(data[i][0],data[i][1]))
    
pos=nx.get_node_attributes(Z1,'pos')
```

```python
def random_edge(graph):

    nonedges = list(nx.non_edges(graph))

    chosen_nonedge = random.choice(nonedges)

    # add new edge
    graph.add_edge(chosen_nonedge[0], chosen_nonedge[1])
```

```python
for _ in range(4414):
    random_edge(Z1)
```

```python
#random.seed(0)
#node_attribute_dict = {node: 0.5+random.random() for node in G1.nodes()}
#nx.set_node_attributes(G1, values=node_attribute_dict, name='expose2infect_weight')

H1 = nx.DiGraph()
H1.add_node('S')
H1.add_edge('E', 'I', rate = 0.7)
H1.add_edge('I', 'R', rate = 0.3)

J1 = nx.DiGraph()
J1.add_edge(('I', 'S'), ('I', 'E'), rate = 0.5)
IC1 = defaultdict(lambda: 'S')
for node in range(10):
    IC1[node] = 'I'
    
return_statuses = ('S', 'E', 'I', 'R')

t, S, E, I, R = EoN.Gillespie_simple_contagion(Z1, H1, J1, IC1, return_statuses,
                                        tmax = float('inf'))

plt.semilogy(t, S, label = 'Susceptible')
plt.semilogy(t, E, label = 'Exposed')
plt.semilogy(t, I, label = 'Infected')
plt.semilogy(t, R, label = 'Recovered')
plt.legend()
```

```python
H2 = nx.DiGraph()
H2.add_node('S')
H2.add_edge('E', 'I', rate = 0.7)
H2.add_edge('I', 'R', rate = 0.3)

J2 = nx.DiGraph()
J2.add_edge(('I', 'S'), ('I', 'E'), rate = 0.5, weight_label='weight')
IC2 = defaultdict(lambda: 'S')
for node in range(10):
    IC2[node] = 'I'
    
return_statuses = ('S', 'E', 'I', 'R')

t, S, E, I, R = EoN.Gillespie_simple_contagion(Z2, H2, J2, IC2, return_statuses,
                                        tmax = float('inf'))

plt.semilogy(t, S, label = 'Susceptible')
plt.semilogy(t, E, label = 'Exposed')
plt.semilogy(t, I, label = 'Infected')
plt.semilogy(t, R, label = 'Recovered')
plt.legend()

plt.savefig('results/ss_5new.png')
```
