```python
import time

from mpl_toolkits.mplot3d import Axes3D
import seaborn as sns
import matplotlib.pyplot as plt
import umap.umap_ as umap
from sklearn.datasets import make_blobs
import numpy as np
%matplotlib inline
import scipy
import time
import networkx as nx
import collections
# import EoN
from collections import defaultdict
import random
import pandas as pd
from seirsplus.models import *
from seirsplus.networks import *
import networkx
```

```python
s = 100
arr = [int(s)]
data,y = make_blobs(n_samples=arr,n_features=2,random_state=16, centers=[[1,1]] ,cluster_std=0.5)
plt.scatter(data[:,0],data[:,1], c=y)
```

```python
data.shape
```

```python
G = nx.Graph()

for i in range(len(data)):
    G.add_node(i,pos=(data[i][0],data[i][1]))
    
pos=nx.get_node_attributes(G,'pos')
```

```python
from sklearn.neighbors import kneighbors_graph
```

```python
A2 = kneighbors_graph(data, 2, mode='connectivity', include_self=False)
```

```python
A5 = kneighbors_graph(data, 5, mode='connectivity', include_self=False)
```

```python
A7 = kneighbors_graph(data, 7, mode='connectivity', include_self=False)
```

```python
G2 = nx.from_scipy_sparse_matrix(A2)
G5 = nx.from_scipy_sparse_matrix(A5)
G7 = nx.from_scipy_sparse_matrix(A7)
```

```python
nx.draw(G2, node_size=10)
```

```python
SIGMA  = 1/5.2
GAMMA  = 1/10
MU_I   = 0.002

R0     = 2.5
BETA   = 1/(1/GAMMA) * R0
BETA_Q = 0.5*BETA

P      = 0.2
Q      = 0.05
```

```python
model_base2 = SEIRSNetworkModel(G = G2, 
                          beta = BETA, 
                          sigma = SIGMA, 
                          gamma = GAMMA, 
                          initE = 1000, initI = 500)

model_base5 = SEIRSNetworkModel(G = G5, 
                          beta = BETA, 
                          sigma = SIGMA, 
                          gamma = GAMMA, 
                          initE = 1000, initI = 500)

model_base7 = SEIRSNetworkModel(G = G7, 
                          beta = BETA, 
                          sigma = SIGMA, 
                          gamma = GAMMA, 
                          initE = 1000, initI = 500)
```

```python
model_base2.run(T=300)
model_base2.figure_basic(ylim=0.2)
```

```python
model_base5.run(T=300)
model_base5.figure_basic(ylim=0.2)
```

```python
model_base7.run(T=300)
model_base7.figure_basic(ylim=0.2)
```
